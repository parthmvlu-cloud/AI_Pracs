Breadth First Search (BFS)

```python
from collections import deque

# Graph with edge costs
graph = {
    "College": [("Western Express Highway", 0.3)],

    "Western Express Highway": [
        ("Guru Hargovindji Road", 0.5),
        ("Chakala", 1.0)
    ],

    "Guru Hargovindji Road": [
        ("JB Nagar", 0.8)
    ],

    "Chakala": [
        ("Home", 0.6)
    ],

    "JB Nagar": [
        ("Home", 0.2)
    ],

    "Home": []
}

queue = deque()
queue.append(("College", ["College"], 0))

visited = []
nodes_explored = 0

while queue:

    current, path, cost = queue.popleft()
    nodes_explored += 1

    if current == "Home":
        print("\n===== Breadth First Search =====")
        print("Path :", " -> ".join(path))
        print("Total Path Cost :", cost, "km")
        print("Nodes Explored :", nodes_explored)
        break

    if current not in visited:
        visited.append(current)

        for neighbour, distance in graph[current]:
            if neighbour not in visited:
                queue.append((neighbour,
                              path + [neighbour],
                              cost + distance))
```

Uniform Cost Search

```python
import heapq

# Graph with edge costs
graph = {
    "College": [("Western Express Highway", 0.3)],

    "Western Express Highway": [
        ("Guru Hargovindji Road", 0.5),
        ("Chakala", 1.0)
    ],

    "Guru Hargovindji Road": [
        ("JB Nagar", 0.8)
    ],

    "Chakala": [
        ("Home", 0.6)
    ],

    "JB Nagar": [
        ("Home", 0.2)
    ],

    "Home": []
}

priority_queue = []
heapq.heappush(priority_queue, (0, "College", ["College"]))

visited = []
nodes_explored = 0

while priority_queue:

    cost, current, path = heapq.heappop(priority_queue)
    nodes_explored += 1

    if current == "Home":
        print("\n===== Uniform Cost Search =====")
        print("Path :", " -> ".join(path))
        print("Total Path Cost :", cost, "km")
        print("Nodes Explored :", nodes_explored)
        break

    if current not in visited:
        visited.append(current)

        for neighbour, distance in graph[current]:
            if neighbour not in visited:
                heapq.heappush(
                    priority_queue,
                    (cost + distance,
                     neighbour,
                     path + [neighbour])
                )
```

Greedy Best-First Search

```python
import heapq

# Graph with edge costs
graph = {
    "College": [("Western Express Highway", 0.3)],

    "Western Express Highway": [
        ("Guru Hargovindji Road", 0.5),
        ("Chakala", 1.0)
    ],

    "Guru Hargovindji Road": [
        ("JB Nagar", 0.8)
    ],

    "Chakala": [
        ("Home", 0.6)
    ],

    "JB Nagar": [
        ("Home", 0.2)
    ],

    "Home": []
}

# Heuristic values
heuristic = {
    "College": 1.8,
    "Western Express Highway": 1.5,
    "Guru Hargovindji Road": 1.0,
    "Chakala": 0.7,
    "JB Nagar": 0.2,
    "Home": 0
}

priority_queue = []
heapq.heappush(priority_queue, (heuristic["College"], "College", ["College"], 0))

visited = []
nodes_explored = 0

while priority_queue:

    h, current, path, cost = heapq.heappop(priority_queue)
    nodes_explored += 1

    if current == "Home":
        print("\n===== Greedy Best-First Search =====")
        print("Path :", " -> ".join(path))
        print("Total Path Cost :", cost, "km")
        print("Nodes Explored :", nodes_explored)
        break

    if current not in visited:
        visited.append(current)

        for neighbour, distance in graph[current]:
            if neighbour not in visited:
                heapq.heappush(
                    priority_queue,
                    (
                        heuristic[neighbour],
                        neighbour,
                        path + [neighbour],
                        cost + distance
                    )
                )
```

A* Search

Formula
f(n) = g(n) + h(n)
g(n) = Cost from College to current node
h(n) = Estimated cost from current node to Home
f(n) = Total estimated cost

```python
import heapq

# Graph with edge costs
graph = {
    "College": [("Western Express Highway", 0.3)],

    "Western Express Highway": [
        ("Guru Hargovindji Road", 0.5),
        ("Chakala", 1.0)
    ],

    "Guru Hargovindji Road": [
        ("JB Nagar", 0.8)
    ],

    "Chakala": [
        ("Home", 0.6)
    ],

    "JB Nagar": [
        ("Home", 0.2)
    ],

    "Home": []
}

# Heuristic values
heuristic = {
    "College": 1.8,
    "Western Express Highway": 1.5,
    "Guru Hargovindji Road": 1.0,
    "Chakala": 0.7,
    "JB Nagar": 0.2,
    "Home": 0
}

priority_queue = []
heapq.heappush(priority_queue, (heuristic["College"], 0, "College", ["College"]))

visited = []
nodes_explored = 0

while priority_queue:

    f, g, current, path = heapq.heappop(priority_queue)
    nodes_explored += 1

    if current == "Home":
        print("\n===== A* Search =====")
        print("Path :", " -> ".join(path))
        print("Total Path Cost :", g, "km")
        print("Nodes Explored :", nodes_explored)
        break

    if current not in visited:
        visited.append(current)

        for neighbour, distance in graph[current]:

            if neighbour not in visited:

                new_g = g + distance
                new_f = new_g + heuristic[neighbour]

                heapq.heappush(
                    priority_queue,
                    (
                        new_f,
                        new_g,
                        neighbour,
                        path + [neighbour]
                    )
                )
```
